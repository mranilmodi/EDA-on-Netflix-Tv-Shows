# Check Duplicate and null values

```python
Netflix_df[Netflix_df.duplicated()]
```

```python
Netflix_df.isnull().sum().sort_values(ascending=False)
```

    Highst missing vlaue in Director,
    next country and cast column

    Lest replace/ filling tha missing value

```python
Netflix_df.director.fillna('No Director', inplace=True)
Netflix_df.cast.fillna('No Cast', inplace=True)
Netflix_df.country.fillna('Country Unavailable', inplace=True)

Netflix_df.dropna(subset=['date_added','rating'],inplace=True)
```

    In this dataset missing value in director, cast and country,

    missing value in Director column filled done with ('No Director')
    missing value in Cast column filled done with ('No Cast')
    missing value in Country column filled with ('No Country')
    and next 2 column date_added, and rating, we must need to fix value for fill missing value but we have
    no any fix value, so drop the missng values in tha Dataset of Netflix

```python
Netflix_df.isnull().sum()
```

#### Let;s we chnage the columns name ( Listed_in) to Genres

```python
Netflix_df.rename(columns = {'listed_in': 'genres'}, inplace = True) 
```

### Add columns month and year from dateadded

```python
Netflix_df['add_month'] = Netflix_df['date_added'].dt.month_name()
Netflix_df['add_year'] = Netflix_df['date_added'].dt.year

```

```python
Netflix_df.drop('date_added', axis = 1, inplace = True)
```

```python
Netflix_df['duration'].str.split(' ',expand=True)
```

## Type

```python
print(Netflix_df['type'].value_counts())
print()
Netflix_df['type'].value_counts().plot(kind = 'pie', autopct = '%.2f')
```

## Release_year

```python
Netflix_df['add_year'].value_counts().sort_values(ascending=False)
```

```python
sns.countplot(data=Netflix_df, x='add_year', 
              order=Netflix_df['add_year'].value_counts(ascending=False).index[0:10],
             hue = Netflix_df['type'])

plt.xlabel('year')
plt.ylabel('Number of Mvies and TV shows')
plt.title('Top 10 year Data of release movies and tv show');
```

## Genres

```python
genrs = Netflix_df.set_index('title').genres.str.split(', ', expand=True).stack().reset_index(level =1 , drop= True)
genrs
```

```python
plt.figure(figsize=(13,10)) 
sns.countplot(y =genrs, order = genrs.value_counts().sort_values(ascending = False).index[:20] )
plt.title('Top 20 Genres on Netflix') 
plt.xlabel('Titles') 
plt.ylabel('Genres') 
```

### Q2: Which year release the maximum number of movies and tv shows in india ? explain by graph

```python
Netlix_india_df = Netflix_df[Netflix_df['country']=='India']
Netlix_india_df2= Netlix_india_df.groupby(['release_year'])['type'].value_counts().sort_values(ascending=False)
Netlix_india_df2.head(10)
```

```python
Netlix_india_df2.head(10).plot(kind='bar');
plt.ylabel('no.of movies')
plt.title('top release year in india on Netflix');
```

### Q3: Which type of Genres popular in india on Netflix ? show with Graph !


```python
Netflix_india_ = Netflix_df[Netflix_df['country']=='India']

Netflix_india_geners = Netflix_india_.set_index('title').genres.str.split(',',expand=True).stack().reset_index(level=1, drop=True)
Netflix_india_geners
```

```python
sns.countplot(y=Netflix_india_geners, order= Netflix_india_geners.value_counts().index[:10])
```

### Q4: In how many movies / Tv shows, Anupam kher was cast ?

```python
Netflix_anupam =  Netflix_df[Netflix_df['cast'].str.contains('Anupam Kher')]

Netflix_anupam_df = Netflix_anupam.set_index('title').type.str.split(',',expand=True).stack().reset_index(level=1, drop=True)

Netflix_anupam_df.value_counts().plot(kind='pie', autopct= '%.2f%%');
plt.title('Contributor of Anupam kher on Netflix');
```
